```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

    def delete_nth_from_end(self, n):
        #Create a dummy node and initialize slow and fast pointers
        dummy = Node(0)
        dummy.next = self.head
        slow = dummy
        fast = dummy

        #Move the fast pointer n+1 steps ahead
        for i in range(n + 1):
            fast = fast.next

        #Move both pointers together until fast reaches the end
        while fast:
            slow = slow.next
            fast = fast.next

        #Delete the N-th node from the end
        slow.next = slow.next.next

    def display(self):
        current = self.head
        elements = []
        while current:
            elements.append(str(current.data))
            current = current.next
        print(", ".join(elements))

#Create a linked list and populate it
my_list = LinkedList()
my_list.append(1)
my_list.append(50)
my_list.append(11)
my_list.append(33)
my_list.append(21)
my_list.append(40)
my_list.append(71)

print("Original Linked List:")
my_list.display()

#Delete the N-th node from the end (N = 2)
n = 2
my_list.delete_nth_from_end(n)
my_list.display()
```
